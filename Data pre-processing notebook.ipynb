```python
import pandas as pd
import re
```

```python
df=pd.read_csv('dialect_with_sentences.csv')
```

```python
df.head()
```

```python
df_copy=df.copy()
```

### Remove everything except arabic characters.

```python
df_copy.sentence=df_copy.sentence.apply(lambda x :re.sub('[^ء-ي ]',' ',x).strip(' '))
```

```python
df_copy.sentence.iloc[0]
```

### Remove multiple spaces

```python
df_copy.sentence=df_copy.sentence.apply(lambda x :re.sub(' +',' ',x))
```

```python
df_copy.sentence.iloc[0]
```

```python
df_copy.info()
```

### I would remove any characters duplicated more than 2 times.
### Buف the  word 'ههههه' and its duplicates would be converted to ههه 

```python
re.sub(r'([^ه])\1\1+', r'\1','مساء الفللللل والياسمين والعمبر ههههههه')
```

```python
re.sub(r'([ه])\1\1+', r'\1\1\1','مساء الفللللل والياسمين والعمبر ههههههه')
```

```python
df_copy.sentence=df_copy.sentence.apply(lambda x :re.sub(r'([^ه])\1\1+', r'\1',x))
```

```python
df_copy.sentence=df_copy.sentence.apply(lambda x :re.sub(r'([ه])\1\1+', r'\1\1\1',x))
```
